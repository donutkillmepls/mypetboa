```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import PolynomialFeatures
import warnings
warnings.filterwarnings("ignore")
np.random.seed(42)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
df=pd.read_csv('https://piazza.com/redirect/s3?bucket=uploads&prefix=paste%2Fjoofj6t8bak2xe%2F91969386d42aa27a910a464b7987a9b225451282fa2934baf2c1afb9ef6480c7%2Floan_data.csv')
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
df['purpose']=label_encoder.fit_transform(df['purpose'])
```

```python
df.purpose.unique
```

```python
df.describe()
```

```python
y=df.drop('not.fully.paid',axis=1)
```

```python
train,test,y_train,y_test=train_test_split(df,y,test_size=0.2)
```

```python
rf=RandomForestClassifier()

```

```python
rf.fit(train,y_train)
```

```python
prediction=rf.predict(test)
```

```python
print(accuracy_score(y_test,prediction))
```

```python
lr=LogisticRegression()
lr.fit(train,y_train)
pred=lr.predict(test)
print(accuracy_score(pred,y_test))
```

```python
dt=DecisionTreeClassifier()
dt.fit(train,y_train)
pred=dt.predict(test)
print(accuracy_score(pred,y_test))
```
